```python
import pandas as pd
import pyperclip

# 데이터프레임 생성 예제
df = pd.read_excel("c:/121.xlsx")

# 'value' 열을 부동소수점(float)으로 변환
values_float = df['value'].astype(float)

# 'value' 열 값을 개행 문자로 구분하여 클립보드에 복사
values_newline_sep = "\n".join(values_float.astype(str))
pyperclip.copy(values_newline_sep)

# 클립보드에 복사된 내용을 붙여넣기 가능
print("클립보드에 복사되었습니다:\n", values_newline_sep)
```
