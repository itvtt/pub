```python
import psutil



# CPU 사용량
print(f"CPU 사용량: {psutil.cpu_percent()}%")

# 메모리 사용량
mem = psutil.virtual_memory()
print(f"전체 메모리: {mem.total / 1024 / 1024:.2f}MB")
print(f"사용 중인 메모리: {mem.used / 1024 / 1024:.2f}MB")
print(f"사용 가능한 메모리: {mem.available / 1024 / 1024:.2f}MB")
print(f"메모리 사용량: {mem.percent}%")

# 디스크 사용량
disk = psutil.disk_usage('/')
print(f"전체 디스크 용량: {disk.total / 1024 / 1024:.2f}MB")
print(f"사용 중인 디스크 용량: {disk.used / 1024 / 1024:.2f}MB")
print(f"사용 가능한 디스크 용량: {disk.free / 1024 / 1024:.2f}MB")
print(f"디스크 사용량: {disk.percent}%")



# 현재 실행 중인 프로세스의 정보를 가져옵니다.
processes = []

for proc in psutil.process_iter(['pid', 'name', 'cpu_percent']):
    try:
        # 프로세스 이름, PID, CPU 사용량을 가져옵니다.
        process = {
            'name': proc.info['name'],
            'pid': proc.info['pid'],
            'cpu_percent': proc.info['cpu_percent']
        }

        # 프로세스 정보를 리스트에 추가합니다.
        processes.append(process)
    except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess):
        pass

# CPU 사용량이 높은 순으로 프로세스를 정렬합니다.
processes = sorted(processes, key=lambda x: x['cpu_percent'], reverse=True)

# 정렬된 프로세스 정보를 출력합니다.
for p in processes:
    print(f"{p['name']} ({p['pid']}): {p['cpu_percent']}%")

# CPU 사용량 측정
cpu_percent = psutil.cpu_percent(interval=0.1)

# 측정한 CPU 사용량 출력
print(f"CPU 사용량: {cpu_percent}%")

```
