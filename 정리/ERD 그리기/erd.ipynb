```python
##1. 차트하나에 그리기

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
data = pd.read_csv('c:/erd.csv')

# 데이터 확인하기
print(data.head())

# 바 차트 그리기
fig, ax = plt.subplots()
bar_width = 0.35

# LSL, USL 범위 표시
ax.bar(data.index, data['USL'] - data['LSL'], bar_width, data['LSL'], color='r', alpha=0.5, label='Range')
ax.set_xlabel('Index')
ax.set_ylabel('Value')
ax.set_title('Range Bar Chart')
ax.legend()

# Value 좌표로 그리기
ax.scatter(data.index, data['Value'], color='b', label='Value')

# X 축 레이블 설정
ax.set_xticks(data.index)
ax.set_xticklabels(data['NAME'])

plt.tight_layout()
plt.show()
```

```python
## 차트를 개별로 나타내기
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
data = pd.read_csv('c:/erd.csv')


# 데이터 확인하기
print(data.head())

# 차트 그리기
fig, axs = plt.subplots(nrows=len(data), ncols=1, figsize=(6, 4*len(data)))

for i, row in data.iterrows():
    name = row['NAME']
    value = row['Value']
    lsl = row['LSL']
    usl = row['USL']

    ax = axs[i]
    ax.bar(name, usl - lsl, color='r', alpha=0.5, label='Range')
    ax.set_xlabel('Index')
    ax.set_ylabel('Value')
    ax.set_title(f'Range Bar Chart - {name}')
    ax.legend()

    ax.scatter(name, value, color='b', label='Value')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
data = pd.read_csv('c:/erd.csv')

# 데이터 확인하기
print(data.head())

# NAME으로 그룹화하여 차트 그리기
fig, axs = plt.subplots(nrows=len(data.groupby('NAME')), ncols=1, figsize=(8, 6*len(data.groupby('NAME'))))

grouped_data = data.groupby('NAME')

for i, (name, group) in enumerate(grouped_data):
    ax = axs[i]
    ax.bar(group['slot'], group['USL'] - group['LSL'], bottom=group['LSL'], color='skyblue', alpha=0.5, label='Range')
    ax.set_xlabel('Slot')
    ax.set_ylabel('Value')
    ax.set_title(f'Range Bar Chart - {name}')
    
    unique_lotids = group['lotid'].unique()
    colors = plt.cm.tab10(range(len(unique_lotids)))  # 색상 맵 생성
    
    for lotid, color in zip(unique_lotids, colors):
        lotid_group = group[group['lotid'] == lotid]
        ax.scatter(lotid_group['slot'], lotid_group['Value'], color=color, label=lotid)
    
    ax.legend()

    ax.set_xticks(range(1, 26, 2))  # X 축 눈금 설정 (간격을 2로 설정)
    ax.set_xticklabels(range(1, 26, 2))  # X 축 눈금 라벨 설정

plt.tight_layout()
plt.show()

# 결과를 엑셀로 저장
result_data = data.copy()
result_data['Range'] = result_data['USL'] - result_data['LSL']
result_data.to_excel('c:/erd_out.xlsx', index=False)
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 불러오기
data = pd.read_csv('c:/erd.csv')

# 데이터 확인하기
print(data.head())

# NAME으로 그룹화하여 차트 그리기
fig, axs = plt.subplots(nrows=len(data.groupby('NAME')), ncols=1, figsize=(8, 6*len(data.groupby('NAME'))))

grouped_data = data.groupby('NAME')

for i, (name, group) in enumerate(grouped_data):
    ax = axs[i]
    
    unique_lotids = group['lotid'].unique()
    num_lotids = len(unique_lotids)
    bar_width = 0.8 / num_lotids  # 각 lotid에 대한 바 너비 계산
    offset = (1 - bar_width) / 2  # 바 시작 위치 계산
    
    for idx, lotid in enumerate(unique_lotids):
        lotid_group = group[group['lotid'] == lotid]
        ax.bar(lotid_group['slot'] + (idx * bar_width) + offset,
               lotid_group['USL'] - lotid_group['LSL'],
               bottom=lotid_group['LSL'],
               width=bar_width,
               color='skyblue',
               alpha=0.5,
               label='Range')
        ax.scatter(lotid_group['slot'] + (idx * bar_width) + offset,
                   lotid_group['Value'],
                   color=plt.cm.tab10(idx),
                   label=lotid)
    
    ax.set_xlabel('Slot')
    ax.set_ylabel('Value')
    ax.set_title(f'Range Bar Chart - {name}')
    ax.legend()
    ax.set_xticks(range(1, 26, 1))  # X 축 눈금 설정 (간격을 2로 설정)
    ax.set_xticklabels(range(1, 26, 1))  # X 축 눈금 라벨 설정

plt.tight_layout()
plt.show()

# 결과를 엑셀로 저장
result_data = data.copy()
result_data['Range'] = result_data['USL'] - result_data['LSL']
result_data.to_excel('c:/erd_out.xlsx', index=False)
```
