```python
# 정확히 일치하게

import csv

filename = 'a.csv'
search_term = '타입'  # 또는 '타입B'

with open(filename, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if search_term in row:
            print(', '.join(row))
```
