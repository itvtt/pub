```python
from PIL import Image

def split_image(image_path):
    image = Image.open(image_path)
    image = image.resize((600, 600))
    width, height = image.size
    half_width = width // 2
    half_height = height // 2

    left_upper = image.crop((0, 0, half_width, half_height))
    left_lower = image.crop((0, half_height, half_width, height))
    right_upper = image.crop((half_width, 0, width, half_height))
    right_lower = image.crop((half_width, half_height, width, height))

    return left_upper, left_lower, right_upper, right_lower

#      .
left_upper, left_lower, right_upper, right_lower = split_image('python.png')


split_image('python.png')
display(left_upper,left_lower,right_upper,right_lower)
```
