### 이미지 자르기

### 가우시안 블러 처리

```python
from PIL import Image
from PIL import ImageFilter
img = Image.open('python.png')
img.filter(ImageFilter.GaussianBlur(4))
```

### 머리 부분만 블러처리해보기

```python
img.save('python_head_blurred.png')
```

```python
from PIL import Image
from PIL import ImageFilter
img = Image.open('python.png')
```

```python
display(img)
```

```python
head = img.crop((0,0,550,550))
```

```python
head_blurred = head.filter(ImageFilter.GaussianBlur(5))
```

```python
img.paste(head_blurred, (0,0,550,550))
```
