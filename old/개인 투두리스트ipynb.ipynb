```python
import tkinter as tk
from tkinter import messagebox
from datetime import datetime

TODO_FILE_PATH = "C:/todo.txt"
COMP_FILE_PATH = "C:/comp.txt"

GROUPS = {
    "그룹1": ["당근", "감자", "토마토"],
    "그룹2": ["피자", "치킨", "햄버거"]
}

def add_todo():
    todo = entry.get()
    group = group_var.get()
    if todo and group:
        listbox.insert(tk.END, f"[{group}] {todo}")
        entry.delete(0, tk.END)
        save_todos()
    else:
        messagebox.showwarning("경고", "할 일과 그룹을 선택하세요.")

def delete_todos():
    selected_indices = listbox.curselection()
    if selected_indices:
        confirmed = messagebox.askyesno("확인", "선택한 투두를 삭제하시겠습니까?")
        if confirmed:
            deleted_todos = []
            for index in reversed(selected_indices):
                deleted_todos.append(listbox.get(index))
                listbox.delete(index)
            save_todos()
            save_completed_todos(deleted_todos)
    else:
        messagebox.showwarning("경고", "할 일을 선택하세요.")

def save_todos():
    todos = listbox.get(0, tk.END)
    with open(TODO_FILE_PATH, "w") as file:
        file.write("\n".join(todos))

def save_completed_todos(deleted_todos):
    with open(COMP_FILE_PATH, "a") as file:
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        for todo in deleted_todos:
            file.write(f"{timestamp}: {todo}\n")

def load_todos():
    try:
        with open(TODO_FILE_PATH, "r") as file:
            todos = file.read().splitlines()
            for todo in todos:
                listbox.insert(tk.END, todo)
    except FileNotFoundError:
        pass

def on_check(event):
    index = listbox.nearest(event.y)
    if index >= 0:
        listbox.selection_clear(0, tk.END)
        listbox.selection_set(index)
        delete_todos()

def on_group_select():
    selected_group = group_var.get()
    if selected_group:
        for child in group_frame.winfo_children():
            child.destroy()

        group_todos = GROUPS[selected_group]
        for todo in group_todos:
            listbox_group.insert(tk.END, todo)

root = tk.Tk()
root.title("투두 리스트")

listbox = tk.Listbox(root, height=10)
listbox.pack(pady=10)

scrollbar = tk.Scrollbar(root)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
listbox.config(yscrollcommand=scrollbar.set)
scrollbar.config(command=listbox.yview)

entry = tk.Entry(root)
entry.pack(pady=10)

group_frame = tk.Frame(root)
group_frame.pack(pady=10)

group_var = tk.StringVar()

for group in GROUPS:
    tk.Radiobutton(root, text=group, variable=group_var, value=group, padx=10, command=on_group_select).pack()

add_button = tk.Button(root, text="추가", command=add_todo)
add_button.pack(pady=5)

delete_button = tk.Button(root, text="삭제", command=delete_todos)
delete_button.pack(pady=5)

listbox.bind("<Button-1>", on_check)

listbox_group = tk.Listbox(group_frame, height=5)
listbox_group.pack(side=tk.LEFT)

load_todos()

root.mainloop()
```

```python
import tkinter as tk

todo_list = []

class TodoItem:
    def __init__(self, group, checklist, description, completed=False):
        self.group = group
        self.checklist = checklist
        self.description = description
        self.completed = completed

def add_todo_item():
    description = entry.get()
    if description:
        group = group_var.get()
        checklist = checklist_var.get()
        todo_item = TodoItem(group, checklist, description)
        todo_list.append(todo_item)
        update_todo_list()

def toggle_completion_status(index):
    if 0 <= index < len(todo_list):
        todo_list[index].completed = not todo_list[index].completed
        update_todo_list()

def update_todo_list():
    todo_frame.destroy()

    todo_frame = tk.Frame(root)
    todo_frame.pack()

    for index, todo_item in enumerate(todo_list):
        checkbox = tk.Checkbutton(todo_frame, command=lambda index=index: toggle_completion_status(index))
        checkbox.grid(row=index, column=0)

        label = tk.Label(todo_frame, text=f"{todo_item.group}호기 체크리스트 {todo_item.checklist}: {todo_item.description}")
        label.grid(row=index, column=1)

# 윈도우 생성
root = tk.Tk()

# 할일 추가 입력 필드
entry = tk.Entry(root)
entry.pack()

# 그룹 선택 라디오 버튼
group_var = tk.StringVar()
group_var.set("1호기")

group_radio_1 = tk.Radiobutton(root, text="1호기", variable=group_var, value="1호기")
group_radio_1.pack()

group_radio_2 = tk.Radiobutton(root, text="2호기", variable=group_var, value="2호기")
group_radio_2.pack()

# 체크리스트 선택 라디오 버튼
checklist_var = tk.StringVar()
checklist_var.set("A")

checklist_radio_a = tk.Radiobutton(root, text="A", variable=checklist_var, value="A")
checklist_radio_a.pack()

checklist_radio_b = tk.Radiobutton(root, text="B", variable=checklist_var, value="B")
checklist_radio_b.pack()

checklist_radio_c = tk.Radiobutton(root, text="C", variable=checklist_var, value="C")
checklist_radio_c.pack()

# 할일 추가 버튼
add_button = tk.Button(root, text="Add", command=add_todo_item)
add_button.pack()

# 할일 목록 프레임
todo_frame = tk.Frame(root)
todo_frame.pack()

# GUI 시작
root.mainloop()
```
