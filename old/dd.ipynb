```python
import matplotlib.pyplot as plt
import numpy as np

# 121개의 랜덤한 3자리 수 생성
np.random.seed(42)  # 랜덤 값 고정을 위한 시드 설정
values = np.random.randint(100, 1000, 121)

# 극 좌표계로 변환
angles = np.linspace(0, 2*np.pi, len(values), endpoint=False)
radii = values
colors = ['blue' if value >= 250 else 'red' for value in values]

# 웨이퍼 맵 시각화
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.scatter(angles, radii, c=colors)
ax.set_rmax(max(values))  # 반지름 최댓값 설정
ax.set_yticklabels([])  # 반지름 축 레이블 제거

# 임계값을 나타내는 선 그리기
ax.plot([0, 2*np.pi], [250, 250], color='gray', linestyle='--', label='Threshold (250)')
ax.legend()

plt.title('Wafer Map Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 121개의 랜덤한 3자리 수 생성
np.random.seed(42)  # 랜덤 값 고정을 위한 시드 설정
values = np.random.randint(100, 1000, 121).reshape(11, 11)

# 원형 웨이퍼 맵 시각화
theta, r = np.meshgrid(np.linspace(0, 2*np.pi, 11), np.linspace(0, 1, 11))
colors = ['blue' if value >= 250 else 'red' for value in values.flatten()]

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
sc = ax.scatter(theta, r, c=colors, s=100, cmap='coolwarm')
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.spines['polar'].set_visible(False)
ax.grid(False)

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Value')
plt.title('Circular Wafer Map')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# 이미지 불러오기
image_path = 'dd.jpg'
image = mpimg.imread(image_path)

# 포인트의 개수 설정
num_points = 121  # 그릴 점의 개수

# 2차원 정규 분포로 포인트 좌표 생성
mean = [image.shape[1] / 2, image.shape[0] / 2]  # 이미지 중심을 평균값으로 설정
cov = [[1000, 0], [0, 1000]]  # 공분산 매트릭스
points = np.random.multivariate_normal(mean, cov, num_points)

# 색상 맵 선택
cmap = plt.get_cmap('viridis')  # viridis 색상 맵 사용

# 이미지 위에 점 그리기
plt.imshow(image)

for i in range(num_points):
    x, y = points[i]
    normalized_x = int(x) % image.shape[1]  # 이미지 범위 내의 값으로 변환
    normalized_y = int(y) % image.shape[0]
    color = cmap(i / num_points)  # 포인트 위치에 따른 색상 선택
    plt.scatter(normalized_x, normalized_y, color=color, s=10)  # 점의 크기 조정

plt.title("Image with Gradient Colored Points")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


center = (0, 0)  # 원의 중심 좌표 (x, y)
radius = 1.0     # 원의 반지름

# 원 그리기
circle = plt.Circle(center, radius, fill=False)
fig, ax = plt.subplots()
ax.add_artist(circle)

# 원 위에 표시할 좌표 리스트 생성
num_points = 10  # 표시할 좌표 개수
angles = np.linspace(0, 2 * np.pi, num_points, endpoint=False)
coordinates = [(center[0] + radius * np.cos(angle), center[1] + radius * np.sin(angle)) for angle in angles]

# 좌표 표시하기
for x, y in coordinates:
    plt.text(x, y, f"({x:.2f}, {y:.2f})", ha='center', va='bottom')

# 그래프 설정
ax.set_aspect('equal', adjustable='box')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
plt.xlabel('X 좌표')
plt.ylabel('Y 좌표')
plt.title('원형 도형 위의 좌표')

# 그래프 표시
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# 원의 중심 좌표와 반지름 입력 받기
center_x = float(input("원의 중심 X 좌표를 입력하세요: "))
center_y = float(input("원의 중심 Y 좌표를 입력하세요: "))
radius = float(input("원의 반지름을 입력하세요: "))

# 원 그리기
circle = plt.Circle((center_x, center_y), radius, fill=False)
fig, ax = plt.subplots()
ax.add_artist(circle)

# 원 위에 표시할 좌표 입력 받기
num_points = int(input("원 위에 표시할 좌표 개수를 입력하세요: "))
coordinates = []
for i in range(num_points):
    x = float(input(f"좌표 {i+1}의 X 좌표를 입력하세요: "))
    y = float(input(f"좌표 {i+1}의 Y 좌표를 입력하세요: "))
    coordinates.append((x, y))

# 좌표 표시하기
for x, y in coordinates:
    plt.text(x, y, f"({x:.2f}, {y:.2f})", ha='center', va='bottom')

# 그래프 설정
ax.set_aspect('equal', adjustable='box')
ax.set_xlim(center_x - radius - 1, center_x + radius + 1)
ax.set_ylim(center_y - radius - 1, center_y + radius + 1)
plt.xlabel('X 좌표')
plt.ylabel('Y 좌표')
plt.title('원형 도형 위의 좌표')

# 그래프 표시
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
data = {
    'point': [1, 2, 3, 4, 5, 6, 7],
    'value': [1, 4, 1, 51, 2, 1525, 163]
}

df = pd.DataFrame(data)

# 원형의 도형 위에 산포도 표시
fig, ax = plt.subplots()
scatter = ax.scatter(df['point'], df['point'], c=df['value'], cmap='viridis', s=100, marker='o')

# 컬러바 설정
cbar = plt.colorbar(scatter)
cbar.set_label('값')

# 그래프 꾸미기
ax.set_title('원형의 도형 위에 값의 정도로 표시')
ax.set_xlabel('포인트')
ax.set_ylabel('포인트')

# 원형의 도형 생성
circle = plt.Circle((0.5, 0.5), 20, color='gray', fill=False)
ax.add_patch(circle)

# 그래프 출력
plt.show()
```
